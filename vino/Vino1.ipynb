```python
import pandas as pd
import matplotlib.pyplot as plt


red_wine = pd.read_csv('winequality-red.csv', delimiter=';')
white_wine = pd.read_csv('winequality-white.csv', delimiter=';')


print(red_wine.head())
print(white_wine.head())
```

```python

wine_data = pd.concat([red_wine, white_wine], ignore_index=True)


wine_data.replace([float('inf'), float('-inf')], float('nan'), inplace=True)


print(wine_data.head())


correlation_matrix = wine_data.corr()


print(correlation_matrix)
```

```python

ph_correlations = correlation_matrix['pH'].drop('pH')
print(ph_correlations)
most_correlated_attribute = ph_correlations.idxmax()
print(f"Svojstvo koje najviše korelira sa vrijednošću ph je: {most_correlated_attribute}")


print(f"Vrijednost korelacija: {ph_correlations[most_correlated_attribute]}")
```

```python

plt.figure(figsize=(10, 5))
plt.hist(wine_data['pH'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram za pH vrijednost')
plt.xlabel('pH')                                                                                   
plt.ylabel('Frekvencija')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.hist(wine_data[most_correlated_attribute].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title(f'Histogram za "{most_correlated_attribute}" što god to bilo')
plt.xlabel(most_correlated_attribute)
plt.ylabel('Frekvencija')
plt.show()
```
